## Data Cleaning and Exploration on City Quality of Life Dataset

```python
import pandas as pd
```

```python
# Load data into df
complete_df = pd.read_csv('uaScoresDataFrame.csv')

# Drop unwanted columns
complete_df = complete_df.drop(['Unnamed: 0'], 1)

# Round numerical data to 3 decimals
complete_df = complete_df.round(3)
complete_df
```

```python
# Drop na values
complete_df.dropna()
```

Note: Dropping na resulted in no change as the dataset we are using had some level of pre-processing already complete
